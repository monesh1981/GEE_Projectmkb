```python
import ee
import geemap
ee.Authenticate()
ee.Initialize(project = 'ee-moneshbathre')
map = geemap.Map()
```

```python
cg_bdy = ee.FeatureCollection("projects/ee-moneshbathre/assets/chattisgarh_bdy")
map.addLayer(cg_bdy,{},"Chattisgarh_boundary")
map.centerObject(cg_bdy,10)
map
```

```python
LULC_cg = ee.ImageCollection("MODIS/061/MCD12C1").select("Majority_Land_Cover_Type_1")
map.addLayer(LULC_cg,{},"CG_LULC")
```

```python
region = ee.Geometry.Polygon(
    [[[80.1945, 17.7844],
    [80.1945, 24.1039],
    [84.4094, 24.1039],
    [84.4094, 17.7844],
    [80.1945, 17.7844]]],
    None,
    False)
```

```python

dataset = ee.ImageCollection("MODIS/061/MCD12C1").select("Majority_Land_Cover_Type_1")
landcover_time_series = dataset.filterDate("2001-01-01", "2023-01-01")

vis_params = {
    "min": 0,
    "max": 16,
    "palette": [
        "1c0dff", "05450a", "086a10", "54a708", "78d203", "009900", "c6b044", "dcd159",
        "dade48", "fbff13", "b6ff05", "27ff87", "c24f44", "a5a5a5", "ff6d4c",
        "69fff8", "f9ffa4"
    ],
}

```

```python
rgbVis = landcover_time_series.map(lambda img: img.visualize( **vis_params).clip(cg_bdy))
```

```python
gifparams ={'region': region,
            'dimensions': 600,
            'crs': 'EPSG:4326',
            'framesPerSecond': 10,
            'format': 'gif'}
```

rgbVis is ImageCollection of visualized NDVI images.
getVideoThumbURL(gifparams) will generate a URL for the animated GIF using the parameters defined in gifparams.

```python
print(rgbVis.getVideoThumbURL(gifparams))
```
