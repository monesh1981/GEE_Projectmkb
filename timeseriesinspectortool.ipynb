```python
import geemap,geemap as map
import ee
ee.Authenticate()
ee.Initialize(project = "ee-moneshbathre")
```

```python
m = geemap.Map()
m
```

```python
region = m.draw_last_feature
region.getInfo()
```

```python
landsat_ts = geemap.landsat_timeseries(roi=region, start_year=1984, end_year=2024, start_date='01-01', end_date='12-31')
```

```python
layer_names= ["Landsat "+str(year) for year in range(1984,2020)]
print(layer_names)
image_dates = landsat_ts.aggregate_array('system:time_start').map(lambda date: ee.Date(date).format('YYYY-MM-dd'))
layer_names = image_dates.getInfo()
```

```python
landsat_viz = {
    'min': 0,
    'max': 0.5,
    'gamma': [1,1,1],
    'bands': ['NIR', 'Red', 'Green']
    }
```

```python
m.ts_inspector(left_ts=landsat_ts, right_ts=landsat_ts, left_names=layer_names, right_names=layer_names, left_vis=landsat_viz, right_vis=landsat_viz)
m.centerObject(region,8)
m
```
