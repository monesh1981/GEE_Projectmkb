```python
import geemap,geemap as map
import ee
ee.Authenticate()
ee.Initialize(project = "ee-moneshbathre")
```

```python
m = map.Map()
m1 = map.Map()
m2 = map.Map()

```

add the shape feature as layer to m m1 and m2
centre the maps

```python
study_area = ee.FeatureCollection("projects/ee-moneshbathre/assets/chennai_bdy")
m.addLayer(study_area,{},"study_area")
m1.addLayer(study_area,{},"study_area")
m2.addLayer(study_area,{},"study_area")

m.centerObject(study_area,10)
m1.centerObject(study_area,10)
m2.centerObject(study_area,10)
m
m1
m2
```

```python
# collect the satellite images from 2024 to 1985 with desired date cloudcover and clipped to study area
landsat5_1985 = ee.ImageCollection("LANDSAT/LM05/C02/T1").filterDate('1985-01-01', '1985-12-31').filterMetadata('CLOUD_COVER', 'less_than',10).mean().clip(study_area)
landsat5_1990 = ee.ImageCollection("LANDSAT/LT05/C02/T1").filterDate('1990-01-01', '1990-12-31').filterMetadata('CLOUD_COVER', 'less_than',10).mean().clip(study_area)
landsat7_2000 = ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2000-01-01', '2000-12-31').filterMetadata('CLOUD_COVER', 'less_than',10).mean().clip(study_area)
landsat7_2010 = ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2010-01-01', '2010-12-31').filterMetadata('CLOUD_COVER', 'less_than',10).mean().clip(study_area)
landsat8_2020 = ee.ImageCollection("LANDSAT/LC08/C02/T1").filterDate('2020-01-01', '2020-12-31').filterMetadata('CLOUD_COVER', 'less_than',10).mean().clip(study_area)
landsat9_2024 = ee.ImageCollection("LANDSAT/LC09/C02/T1").filterDate('2024-01-01', '2024-12-31').filterMetadata('CLOUD_COVER', 'less_than',10).mean().clip(study_area)


```

```python
## define variables to store NDVI maps of respective years
ndvi_2024 = landsat9_2024.normalizedDifference(['B5','B4'])
ndvi_2020 = landsat8_2020.normalizedDifference(['B5','B4'])
ndvi_2010 = landsat7_2010.normalizedDifference(['B4','B3'])
ndvi_2000 = landsat7_2000.normalizedDifference(['B4','B3'])
ndvi_1990 = landsat5_1990.normalizedDifference(['B3','B2'])
ndvi_1985 = landsat5_1985.normalizedDifference(['B3','B2'])

# add the indices to map layer


m.addLayer(ndvi_2024,{'min':-0.1,'max':0.4,'palette': palette_ndvi},'ndvi_2024')
m.addLayer(ndvi_2020,{'min':-0.1,'max':0.3,'palette':palette_ndvi},'ndvi_2020')
m.addLayer(ndvi_2010,{'min':-0.4,'max':0,'palette':palette_ndvi},'ndvi_2010')
m.addLayer(ndvi_2000,{'min':-0.5,'max':1,'palette':palette_ndvi},'ndvi_2000')
m.addLayer(ndvi_1990,{'min':-0.2,'max':0.1,'palette':palette_ndvi},'ndvi_1990')
m.addLayer(ndvi_1985,{'min':-1,'max':1,'palette':palette_ndvi},'ndvi_1985')
m.addLayerControl()
m
```

```python
#define the palette for display NDWI visualization
ndwi_palette=(['#CEFFFD', '#76E3DE', '#76CCE3', '#40B4D3', '#4078D3', '#2126BB'])
```

```python
# define variables to store NDWI maps of respective years
ndwi_2024 = landsat9_2024.normalizedDifference(['B3','B5'])
ndwi_2020 = landsat8_2020.normalizedDifference(['B3','B5'])
ndwi_2010 = landsat7_2010.normalizedDifference(['B2','B4'])
ndwi_2000 = landsat7_2000.normalizedDifference(['B2','B4'])
ndwi_1990 = landsat5_1990.normalizedDifference(['B2','B4'])
ndwi_1985 = landsat5_1985.normalizedDifference(['B2','B4'])

```

```python
#define the palette for display SAVI visualization
palette_savi = (['#5A2C0F', '#6A3A1C', '#7B4C28', '#8C5E34', '#9D7040', '#AE8250', '#BF9360', '#D1A472', '#E2B684', '#F3C896', '#F7DAA8', '#F9E1B6', '#FBE9C4', '#FFEECC', '#FFF4D2', '#F7F8B4', '#E5F6A6', '#D2F599', '#C0F38B', '#AEF17E', '#9DEF71', '#88EA66', '#74E65B', '#60E250', '#4EDD47', '#3DD93E', '#30CC3A', '#28B532', '#209E2B', '#188825', '#117320', '#0A5E1A', '#064A14', '#00360F',])
```

```python
# define variables to store SAVI maps of respective years
savi_2024 = landsat9_2024.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat9_2024.select('B5'),'RED': landsat9_2024.select('B4')})
savi_2020 = landsat8_2020.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat8_2020.select('B5'),'RED': landsat8_2020.select('B4')})
savi_2010 = landsat7_2010.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat7_2010.select('B4'),'RED': landsat7_2000.select('B3')})
savi_2000 = landsat7_2000.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat7_2000.select('B4'),'RED': landsat7_2000.select('B3')})
savi_1990 = landsat5_1990.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat5_1990.select('B4'),'RED': landsat5_1990.select('B3')})
savi_1985 = landsat5_1985.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat5_1985.select('B4'),'RED': landsat5_1985.select('B3')})

```

```python
# add the indices to map layer m2
m2.addLayer(savi_2024,{'min':-0.1,'max': 0.5,'palette':palette_savi},'savi_2024')
m2.addLayer(savi_2020,{'min':-0.1,'max': 0.4,'palette':palette_savi},'savi_2020')
m2.addLayer(savi_2010,{'min':-0.8,'max': 0.1,'palette':palette_savi},'savi_2010')
m2.addLayer(savi_2000,{'min':-0.7,'max': 0.2,'palette':palette_savi},'savi_2000')
m2.addLayer(savi_1990,{'min':-0.4,'max': 0.2,'palette':palette_savi},'savi_1990')
m2.addLayer(savi_1985,{'min':-1,'max': 1,'palette':palette_savi},'savi_1985')
m2.addLayerControl()
m2
```
