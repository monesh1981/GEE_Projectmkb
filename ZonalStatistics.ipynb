```python
import geemap,geemap as map
import ee
ee.Authenticate()
ee.Initialize(project = "ee-moneshbathre")
```

```python

m = geemap.Map()
m
```

```python
study_area = ee.FeatureCollection("projects/ee-moneshbathre/assets/India_districts")
m.addLayer(study_area,{},"study_area")

m
```

```python
India_dem = ee.Image("USGS/SRTMGL1_003").clip(study_area)
viz = {"min":0,"max":8700,'palette': ["#003300", "#006600",  "#009933",  "#33CC33", "#66FF66", "#E5FFCC",  "#CC9900",  "#996633",  "#804000",  "#662A00",  "#B2B2B2", "#D8D8D8", "#F0F0F0",  "#F5F5F5",  "#FFFFFF" ]}
m.addLayer(India_dem,viz,"India_dem")
m
```

```python
tree_cover = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2019-01-01","2019-12-31").select("Percent_Tree_Cover").mean().clip(study_area)
viz_tree = {"min":0,"max":78,'palette':['bbe029', '0a9501', '074b03']}
m.addLayer(tree_cover,viz_tree,"tree_cover")
m

```

```python
import os
```

```python
out_dir = os.path.join(os.path.expanduser("~"),"Downloads")
India_treecover = os.path.join(out_dir,"India_treecover.csv")

if not os.path.exists(out_dir):
    os.makedirs(out_dir)

geemap.zonal_stats(tree_cover,study_area,India_treecover,stat_type= "MEAN",scale=250)
```
