```python
import geemap,geemap as map
import ee
ee.Authenticate()
ee.Initialize(project = "ee-moneshbathre")
```

```python
m = geemap.Map()
m
```

```python
study_area = ee.FeatureCollection("projects/ee-moneshbathre/assets/India_districts")
m.addLayer(study_area,{},"study_area")
m
```

```python
leafarea_index = ee.ImageCollection("MODIS/061/MCD15A3H").filterDate("2022-01-01","2024-12-31").select("Lai").mean().clip(study_area)
viz_leafarea = {"min":1.2,"max":47.8,"palette":['e1e4b4', '999d60', '2ec409', '0a4b06']}
m.addLayer(leafarea_index,viz_leafarea,"leafarea_index")
m
```

```python
import os
```

```python
out_dir = os.path.join(os.path.expanduser("~"),"Downloads")
India_leafareaindex = os.path.join(out_dir,"India_leafareaindex.csv")

if not os.path.exists(out_dir):
    os.makedirs(out_dir)

geemap.zonal_stats(leafarea_index,study_area,India_leafareaindex,stat_type="MEAN",scale=250)
```
